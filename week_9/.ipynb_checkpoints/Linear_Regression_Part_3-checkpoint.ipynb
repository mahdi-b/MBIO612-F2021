```r
fish_data = read_csv("data/fish_info.csv")
```

### Including Species for the Weight Prediction

* Recall that by default, each level of Species is intepreted as an explanatory variable
  * In the absence of other explanatory variable, the mean of each level is used to explain the reponse.
* Typically visualzed using boxplots
* Combining the species and lenght can be thought of separate linear regressions each for Length from each species
  * This is often referred to parallel slopes regression
  
* Effectively, where are generating are parallel lines (same slope) but with different intercepts

```r
lm_species = lm(Weight ~ Species + 0, fish_data)
lm_length = lm(Weight ~ Length + 0, fish_data)
```

```r
ggplot(fish_data, aes(Species, Weight)) + 
  geom_boxplot() + 
  stat_summary(fun = mean)
```

```r
lm_speicies_length = lm(Weight ~ Length + Species + 0, fish_data)
```

```r
library(moderndive)
ggplot(fish_data, aes(Length, Weight, color=Species)) + 
  geom_point() + 
  geom_parallel_slopes(se=FALSE)
```

### Assessing the Model Fit

* Done the same way as with simple linear regression 
  * inspect the residuals standard error
  * Evaluate the fit based on the coefficient of determination (R-squared)


```r
summary(lm_species)$r.squared
summary(lm_length)$r.squared
summary(lm_speicies_length)$r.squared
```

```r
summary(lm_species)$adj.r.squared
summary(lm_length)$adj.r.squared
summary(lm_speicies_length)$adj.r.squared
```

```r
summary(lm_species)$sigma
summary(lm_length)$sigma
summary(lm_speicies_length)$sigma
```

### Findig the best Explanatory Variables

* One of the objectives is to identify the parameters that best explain the data.
  * This means we are looking for the best subset of explanatory variables among all the explanatory variables
* Why not keep all exlanatory variables?
  * A subset is easier to interpret
  * Precision of predicted intervals and confidence bands will be smaller
  * The more parameters the more we overfit the data
* Two goor criteria to evalutate models with different params are:
  * R^2 covered
  * AIC (Akaike Information Criteria)

### Akaike Information Criteria

* we can use The AIC to find the most optimal model

$$
AIC = n\times log(SS_{residual}) + 2(p+1) - nlog(n)
$$

* first part measure goodness of fit
* Second part penalizes number of parameters in the model. 

* Strategy: compute AIC for each possible combination of explanatory variables
  * Model with the smallest AIC is chosen as the most optimal model. 
* In reality we apply a greedy approach (Forward or backward) and use the AIC to add or remove a variable 
   * 
 

### Akaike Information Criteria - Cont'd

* Total of 2p models that contain subsets of p variables. 
  * $p = 30$, there are $p^30 = 1,073,741,824$ total models

* In reality we apply a greedy approach (Forward or backward) and use the AIC to add or remove a variable 
   * 

* Forward selection: fit $p$ simple linear regressions and add to the null model the variable that results in the lowest AIC. This approach is continued until some AIC incrases.

* Backward selection. We start with all variables in the model, and backward
selection remove the variable with the largest 
 * continue until a stopping rule is reached. 

* Mixed selection: combination of both. Add and remove as needed

* Can be implemented using `stepAIC` in the `MASS` library
  * left for the practical


### Variable Interaction

* Our initial model

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \epsilon 
$$

* increasing $X_1$ by 1 unit, leads to a $\beta_1$ increase in $Y$
* What if inceasing $X_1$ leads to an increase in $X_2$?
  * increasing $X_1$ by 1 unit, leads to increase in X_2 by alpha
    * this lead to increase  $Y$ of $\beta_1 + \alpha \beta_2$
    
* Interaction betwen input variables can be accounted for in model formula



### Variable Interaction _ cont'd 
$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_1 X_2+ \epsilon 
$$

* The hierarchical principle: if we include an interaction in a model, we should also include the main effects, even if the p-values associated with their coefficients are not significant.
